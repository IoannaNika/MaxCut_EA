```python
import pandas as pd
```

```python
instance_types = ["setA", "setC", "setD"]
instance_sizes = {"setA": ['n0000006i05', 'n0000012i05', 'n0000025i05', 'n0000050i05', 'n0000100i05'],
                  "setC": ['n0000006i05', 'n0000012i05', 'n0000025i05', 'n0000050i05', 'n0000100i05'],
                  "setD": ['n0000010i05', 'n0000020i05', 'n0000040i05', 'n0000080i05', 'n0000160i05']}
population_sizes = [8, 16, 32, 64, 128, 256]

all_crossovers = ["CustomCrossover", "CliqueCrossover", "CustomCliqueCrossover", "KMeansCrossover","UniformCrossover", "OnePointCrossover", "TwoPointCrossover"]
crossovers = {"setA": {"CustomCrossover","UniformCrossover", "OnePointCrossover", "TwoPointCrossover"},
              "setC": {"CustomCrossover", "KMeansCrossover","UniformCrossover", "OnePointCrossover", "TwoPointCrossover"},
              "setD": {"CustomCrossover", "CliqueCrossover", "CustomCliqueCrossover", "UniformCrossover", "OnePointCrossover", "TwoPointCrossover"}}

for instance in instance_types:
    print("\n---", instance)
    
    # One colum is an instance size
    for instance_size in instance_sizes[instance]:
        print("-----", instance_size)
        with open("maxcut-instances/" + instance + "/" + instance_size + ".bkv") as f:
            vtr = float(f.read())
        
        # For each instance size and crossover, find the best population size per crossover
        for crossover in all_crossovers:
            if crossover in crossovers[instance]:
                best_error =  101.0
                best_population_size = -1 
                
                for population_size in population_sizes:
                    df = pd.read_pickle("results/" + instance + "/" + instance + "_" + instance_size + "_" + str(population_size) + "_10.pkl")
                    filtered_df = df[df["crossover"] == crossover]
                    best_fitness = filtered_df.loc[filtered_df['crossover'] == crossover].groupby("round")['fitness'].max().median()
                    relative_error = abs(best_fitness - vtr) / vtr
                    
                    if relative_error < best_error:
                        best_error = relative_error
                        best_population_size = population_size
                
                print(round(best_error, 3), "(" + str(best_population_size) + ")")
                
            else:
                print("-")
```
